```python
from google.colab import drive
from google.colab import files

drive.mount('/content/drive')
%cd /content/drive/MyDrive/


%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d, Axes3D
from matplotlib import cm
import itertools

datafile = 'housing_prices.txt'
cols = np.loadtxt(datafile,delimiter=',',usecols=(0,1),unpack=True)
X = np.transpose(np.array(cols[:-1]))
y = np.transpose(np.array(cols[-1:]))
m = y.size
X = np.insert(X,0,1,axis=1)

plt.figure(figsize=(10,6))
plt.plot(X[:,1],y[:,0],'b+',markersize=12)
plt.xlabel('Population of City in 10,000s')
plt.ylabel('Price in $10,000s')
```

```python
def h(w,X):
    return np.dot(X,w)
```

```python
def mse(w,X,y):
    # Ensure the result of the dot product is treated as a scalar
    return float((1./(2*m)) * np.dot((h(w,X)-y).T,(h(w,X)-y))[0,0])
```

```python
def gradient_descend(X, y, w_start = np.zeros(2), alpha=0.01, iterations=200, batch_size = 20):
    w = w_start
    J_values = []
    w_store = []
    m = X.shape[0]
    for i in range(iterations):
        rand = np.random.permutation(m)
        X_shuff = X[rand]
        y_shuff = y[rand]
        for j in range(0, m, batch_size):
            X_batch = X_shuff[j:j + batch_size]
            y_batch = y_shuff[j:j + batch_size]
            m_batch = X_batch.shape[0]
            temp_w = w
            gradient = (1./m_batch) * np.dot((h(w, X_batch) - y_batch).T, X_batch).T
            w = w - alpha * gradient
        J_values.append(mse(w, X, y))
        w_store.append(list(w[:,0]))

    return w, w_store, J_values
```

```python
def predict(xval, w):
    return w[0] + w[1]*xval
```

```python
iterations = 200
alpha = 0.01 # learning rate

w_start = np.zeros((X.shape[1],1))
w, w_store, J_values = gradient_descend(X, y, w_start=w_start)
```

```python
batch_sizes = [1, 5, 10, 20]
J_values_dict = {}
w_dict = {}

for batch_size in batch_sizes:
    w_start = np.zeros((X.shape[1],1))
    w, w_store, J_values = gradient_descend(X, y, w_start=w_start, batch_size=batch_size, alpha=0.01, iterations=200)
    J_values_dict[batch_size] = J_values
    w_dict[batch_size] = w

plt.figure(figsize=(10,6))
for batch_size, J_values in J_values_dict.items():
    plt.plot(range(len(J_values)), J_values, label=f'Batch Size: {batch_size}')

plt.title("Convergence of Cost Function for Different Batch Sizes")
plt.xlabel("Iteration number")
plt.ylabel("Cost function")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(X[:,1],y[:,0],'rx',markersize=10,label='Training Data')

w_batch_20 = w_dict[20]

plt.plot(X[:,1],predict(X[:,1], w_batch_20),'b-',label = 'Batch Size 20: yhat = h(x; w) = %0.2f + %0.2fx'%(w_batch_20[0,0],w_batch_20[1,0]))

plt.ylabel('y')
plt.xlabel('x')
plt.legend()
plt.title('Prediction for Batch Size 20')
plt.show()
```

for population size 160,000, the predicted house price with mini-batch gradient descent with batch size 20 is $155,000

```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

df = load_breast_cancer()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt

X = df.data
y = df.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
lr = LogisticRegression(solver='lbfgs', max_iter=5000)
rfe = RFE(lr, n_features_to_select=2)
rfe = rfe.fit(X_train, y_train)
feature_indices = np.array(range(X.shape[1]))
feature_indices = feature_indices[rfe.support_]
plt.scatter(X_train[:,feature_indices[0]], X_train[:,feature_indices[1]], c =y_train)
plt.xlabel(df.feature_names[feature_indices[0]])
plt.ylabel(df.feature_names[feature_indices[1]])
print('The best features are : ', np.array(df.feature_names)[rfe.support_])
```
